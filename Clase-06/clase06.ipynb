# 1. REDES CONVOLUCIONALES

las redes neuronales son un subconjunto del machine learning y el núcleo de los algoritmos de aprendizaje profundo. Están compuestas por capas de nodos, que incluyen una capa de entrada, una o más capas ocultas y una capa de salida. Cada nodo está conectado a otro y tiene un peso y un umbral asociados. Si la salida de un nodo individual está por encima del valor de umbral especificado, ese nodo se activa y envía datos a la siguiente capa de la red. De lo contrario, no se pasa ningún dato a la siguiente capa de la red.


Las redes neuronales convolucionales (ConvNets o CNN) se usan con mayor frecuencia para tareas de clasificacion y computer vision. 
 Antes de las CNN, se utilizaban métodos manuales de extracción de características que requerían mucho tiempo para identificar objetos en imágenes. Sin embargo, actualmente las redes neuronales convolucionales proporcionan un enfoque más escalable para las tareas de clasificación de imágenes y reconocimiento de objetos al aprovechar los principios del álgebra lineal, en concreto la multiplicación de matrices, para identificar patrones en una imagen. No obstante, estas redes pueden exigir un uso intensivo de recursos informáticos y requerir unidades de procesamiento gráfico (GPU) para entrenar los modelos.


### 1.1. ¿ Como funcionan las redes neuronales ?

Las redes neuronales convolucionales tienen un rendimiento superior con las entradas de **imagen voz o señales de audio**. Las redes neuronales convolucionales se componen de 3 tipos de capas:

- Capa convolucional
- Capa de agrupación
- Capa totalmente conectada



La capa convolucional es la primera capa de una red convolucional. Las primeras capas se centran en identificar caracterisiticas simples, como colores y bordes y a medida que avanzan en capas se comienza a reconocer elementos o formas más grandes que finalmente ayudan a la predicción.

### 1.1.1 Capas convolucionales
La capa de convolución es una operación muy importante en las redes neuronales, ya que es el bloque donde se realizan la mayoria de calculos. Los componentes que requeire son:
 - Datos de entrada
 - filtro
 - mapa de caracteristicas
Como ejemplo podemos abordar un problema muy común a resolver con imagenes digitales. Como primer paso definiremos que es una imagen digital.

#### 1.1.1.1 Imagen digital
Son imágenes que se obtienen a través de un dispositivo de conversión análogico-digital como un scáner, una cámara fotográfica digital o directamente desde el ordenador utilizando cualquier programa de tratamiento de imagenes.

Mediante esta captura se generan unos bits de 1 y 0 que son representativos de la imagen.

#### 1.1.1.2 El píxel
El píxel es una capa básica, en ella se almacenan los valores RGB del color de cada pixel. El nombre viene de la abreviacion **picture** and **element**. Como conclusión los píxeles son unidades de color que componen una imagen.

#### 1.1.1.3 Profundidad del píxel
Una imagen digital está formada por una matriz de píxeles **(a,b,c)** donde **a** representa anchura y **b** altura, el termino **c** representa profundidad lo que da a la imagen una dimensión en 3d y esta será la profundidad del color, es decir, esto es lo que nos permite tener colores distintos en una imagen.<br>

La profundidad vendría a ser una unidad de medida binaria porque cada píxel se forma por bits.<br>

Por ejemplo cuando la profundidad del píxel es 1 bit esta solo contiene dos colores/dos niveles de colores que son dos niveles de gris representados por los valores 0 y 1, esto nos da como resultado un color blanco o negro. Para un profundidad de **8 bits** permite que cada píxel pueda tener 256 colores distintos, es decir, 256 niveles de grises distintos y si la profunididad es de 24 bits podemos llegar a 16 millones de píxeles distintos. <br>

### 1.1.2 Capas agrupadas

La agrupación de capas, también conocida como submuestreo, permite reducir la dimensión mediante la reducción del número de parámetros de la entrada. De manera similar a la capa convolucional, la operación de agrupación barre toda la entrada con un filtro, pero la diferencia es que este filtro no tiene ningún peso. En su lugar, el kernel aplica una función de agregación a los valores dentro del campo receptivo y llena así la matriz de salida. Hay dos tipos principales de agrupación: <br>

- **Agrupación máxima:** a medida que el filtro recorre la entrada, selecciona el píxel con el valor más alto para enviarlo a la matriz de salida. Este enfoque suele utilizarse más que la agrupación media.<br>
- **Agrupación media:** conforme el filtro avanza por la entrada, calcula el valor promedio dentro del campo receptivo para enviarlo a la matriz de salida.<br>

Aunque se pierde mucha información en la capa de agrupación, esta tiene una serie de beneficios para la CNN. Ayuda a reducir la complejidad, mejora la eficiencia y limita el riesgo de sobreajuste. <br>



### 1.1.3 Capas totalmente conectadas
El nombre de la capa totalmente conectada describe con precisión la capa en sí. Como se ha mencionado anteriormente, los valores de píxel de la imagen de entrada no están conectados directamente con la capa de salida en las capas parcialmente conectadas. Sin embargo, en la capa totalmente conectada, cada nodo de la capa de salida sí está conectado directamente a un nodo de la capa anterior.

Esta capa realiza la tarea de clasificación basándose en las características extraídas de las capas anteriores y sus diferentes filtros. Las capas convolucionales y de agrupación suelen utilizar funciones ReLU, mientras que las capas totalmente conectadas generalmente usan una función de activación softmax para clasificar adecuadamente las entradas y generar una probabilidad entre 0 y 1.

### 1.1.3 Función de activación softmax

```python
# Paso 1: Importar las bibliotecas necesarias

import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Paso 2: Cargar y preparar los datos
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalizar las imágenes y expandir dimensiones para los canales de color
x_train = np.expand_dims(x_train, axis=-1) / 255.0
x_test = np.expand_dims(x_test, axis=-1) / 255.0

# Convertir etiquetas a one-hot encoding
y_train = tf.keras.utils.to_categorical(y_train, num_classes=10)
y_test = tf.keras.utils.to_categorical(y_test, num_classes=10)

# Paso 3: Construir el modelo de red neuronal convolucional
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

# Paso 4: Compilar el modelo
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Paso 5: Entrenar el modelo
history = model.fit(x_train, y_train, epochs=5, batch_size=64, validation_split=0.1)

# Paso 6: Evaluar el modelo en el conjunto de prueba
loss, accuracy = model.evaluate(x_test, y_test)
print(f'Loss: {loss:.4f}, Accuracy: {accuracy:.4f}')
```

**categorical_crossentropy** es adecuada cuando las etiquetas de salida se encuentran en formato one-hot encoding (codificación uno-a-uno) y hay múltiples clases

**batch_size** En lugar de actualizar los pesos después de cada ejemplo individual, se hace después de un lote de ejemplos. Usar lotes en lugar de ejemplos individuales puede mejorar la eficiencia del entrenamiento, ya que los cálculos pueden ser paralelizados en hardware como GPUs. En este caso, batch_size=64 significa que se utilizan 64 ejemplos para calcular las actualizaciones de los pesos en cada iteración

**validation_split** se utiliza para dividir automáticamente tus datos de entrenamiento en un subconjunto de entrenamiento y un subconjunto de validación durante el entrenamiento.

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 5))
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(x_train[i], cmap='gray')
    plt.title(f'Label: {i}')
    plt.axis('off')
plt.show()
```

```python
# Graficar la precisión
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Accuracy Over Epochs')
plt.legend()

plt.tight_layout()
plt.show()
```

Algunas observaciones clave al leer la gráfica de precisión:

Si las curvas de precisión de entrenamiento y validación siguen una tendencia similar y aumentan a lo largo de las épocas, es un signo positivo de que el modelo está aprendiendo de manera efectiva tanto los datos de entrenamiento como los datos nuevos.<br>

Si la curva de precisión de entrenamiento sigue aumentando mientras la curva de precisión de validación comienza a disminuir o se estanca, es posible que el modelo esté sobreajustando los datos de entrenamiento y no esté generalizando bien a datos nuevos.<br>

Si la curva de precisión de validación se mantiene más baja que la de entrenamiento, es normal, ya que es más difícil para el modelo generalizar a datos nuevos que ya ha visto.<br>

Si la curva de precisión de validación aumenta significativamente mientras la curva de precisión de entrenamiento se mantiene baja, podría indicar que el modelo no está aprendiendo lo suficientemente bien de los datos de entrenamiento.<br>

```python
# Graficar la función de pérdida
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss Over Epochs')
plt.legend()
```

Si las curvas de pérdida de entrenamiento y validación siguen una tendencia similar y disminuyen a lo largo de las épocas, es un signo positivo de que el modelo está aprendiendo de manera efectiva tanto los datos de entrenamiento como los datos nuevos.

Si la curva de pérdida de entrenamiento sigue disminuyendo mientras la curva de pérdida de validación comienza a aumentar, es posible que el modelo esté sobreajustando los datos de entrenamiento y no esté generalizando bien a datos nuevos.

Si la curva de pérdida de validación se mantiene más alta que la de entrenamiento, es normal, ya que es más difícil para el modelo generalizar a datos nuevos que ya ha visto.

Si la curva de pérdida de validación disminuye significativamente mientras la curva de pérdida de entrenamiento se mantiene alta, podría indicar que el modelo no está aprendiendo lo suficientemente bien de los datos de entrenamiento.

Qué pasaría, por ejemplo, si deseamos realizar una predicción sobre una base de datos dependiente del tiempo, como el mercádo bursatil, los sistemas de conducción autómatas, las predicciones del tiempo en una región, etc. 

Las **redes neuronales recurrentes** (RNN - Recurrent Neural Networks), nos proporsiona una serie de algoritmos potentes para abordar este tipo de problemas. De forma general, este tipo de redes pueden aplicarse en secuencias de longitud arbitraria. 

Alguna de las principales aplicaciones de las RNN son

- Predicciones del mercado bursatil, con el fin de saber cuando invertir.
- Predicción de trayectorias en conducción autómata.
- Procesamiento de lenguajes naturales: reconocimiento de voz, traducción automática, transcripción/traducción automática de voz a texto.
- Entre otras.

Debemos tener en cuenta que las RNN presentan los siguientes problemas:

- Gradientes inestables: que pueden tratarse, entre otra cosas, con dropout recurrente o normalización por capas recurrentes.
- Memoria a corto plazo limitada, que puede ampliarse mediante el uso de unidades LSTM o GRU.

Se distinguen por su "memoria", ya que obtienen información de entradas anteriores para influir en la entrada y salida actuales

Los problemas de las RNN ocurren por el tamaño por el tamaño del gradiente, que es la inclinación de la función de pérdida a lo largo de la curva de error. Cuando el gradiente es demasiado pequeño, continúa haciéndose más pequeño, actualizando los parámetros de peso hasta que se vuelven insignificantes, es decir, 0. Cuando ocurre esto, el algoritmo ya no está aprendiendo. Los gradientes explosivos ocurren cuando el gradiente es demasiado grande y se crea un modelo inestable. En este caso, los pesos del modelo crecerán demasiado y eventualmente se representarán como NaN. Una solución a estos problemas es reducir el número de capas ocultas dentro de la red neuronal, eliminando parte de la complejidad en el modelo de RNN.
