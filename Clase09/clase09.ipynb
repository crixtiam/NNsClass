```python
import random
import numpy as np
import pandas as pd
from tqdm import  tqdm
import os
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
link = "./Dataset/train"

class_cat_wild = os.listdir(link)
```

```python

def read_image(label:str, pathImg:list) -> list:
    list_image = []
    list_labels = []
    for img in tqdm(os.listdir(pathImg)):
        path = pathImg + "/" + img
        path_file = os.path.splitext(path)
        if path_file[1] == ".jpg" or path_file[1] == ".png":          
            image  = cv.imread(path,cv.IMREAD_COLOR)
            image = cv.cvtColor(image,cv.COLOR_BGR2RGB)
            list_image.append(image)
            list_labels.append(label)
    return [list_image,list_labels]

```

```python
img_n,label_n = read_image(class_cat_wild[0],path_img[0])
```
