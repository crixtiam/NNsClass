# **1. Que es inteligencia artificial**

[[Contenidos]](#contents)

[UNESCO – La nueva normalidad](https://www.youtube.com/watch?v=fGMSzSumex4)

La búsqueda de replicar la inteligencia humana en máquinas ha sido una aspiración constante a lo largo de la historia de la humanidad. Desde los albores de la civilización, hemos anhelado dotar a nuestras creaciones con la capacidad de pensar, aprender y resolver problemas de manera similar a como lo hacemos nosotros. Esta búsqueda ha dado lugar a lo que hoy conocemos como inteligencia artificial (IA), un campo multidisciplinario que ha evolucionado de manera extraordinaria, pasando por múltiples etapas y desafíos a lo largo de los siglos.

# **2. Aplicaciones IA**

[[Contenidos]](#contents)

<div align ="left">
  <p>
      La inteligencia artificial ha supuesto un irrupción en muchos ámbitos laborales en los cuales se espera un futuro prometedor.
  </p>

  <ul>
      <li><b>Asistentes virtuales:</b> Gracias a los avances de la IA conversacional dispositivos, chat bots y aplicaciones cuentan con el don de la palabra empleando lenguaje natural. La IA conversacional que es usada por los asistentes virtuales emplea aprendizaje automático y procesamiento de lenguaje natural para recopilar y análizar lenguajes y comprenderlos y generar respuestas similares a las humanas y con ello logrando una interacción de comunicación. </li>
      <li><b>Análisis cognitivo:</b> La analítica cognitiva es una rama muy ambiciosa de estudio para la inteligencia artificial debido a que el proposito de esta es diseñar maquinas que mediante una combinación de algoritmos sean capaces de entender, reconocer, aprender y reproducir el funcionamiento de la mente humana.<br>
      El cambio de paradigma en esta aplicación es generar una ayuda mutua en los trabajos cotidianos del ser humano, es decir, se deja de ver este instrumento como una simple maquina de trabajo y se convierte en un compañero más el cual brindará un apoyo laboral codo a codo en las tareas cotidianas.</li>
      <li><b>Reconocimiento de patrones: </b>
      Una aplicación relevante en este campo consiste en la extracción de información mediante datos y patrones para identificar comportamientos de interes en las personas o sistemas, con el fin de predecir un comportamiento a futuro y con ello tomar las mejores decisiones.
      </li>
      <li><b>Sistemas de aprendizaje: </b>
      Los sistemas de aprendizajes en la educación se pueden automátizar mediante la IA para los cual se desgranan infinidad de oportunidades de libre acceso con una educación equitativa y de calidad. Uno de los grandes usos para la educación pueden ser:
      <br>
      Sistemas de tutorias inteligentes, aprendizaje exploratorio, aprendizaje de idiomas y escritura, entre otros.
      </li>
      <li><b>Robotica: </b>Los proceso robóticso usan algoritmos de IA y
       métodos que imitan y automatizan tareas humanas para apoyar en los procesos corporativos. Es una solución que permite aprovechar al máximo el talento meramente humano y mover a los empleados a posiciones más estratégicas y creativas. </li>
  </ul>

# **6. Que es y como funciona IA**

[[Contenidos]](#contents)


<div align ="left">
  <p>
   La IA es un campo científico de la informática en donde se crean programas y algoritmos que pretenden simular la inteligencia y los comportamientos propios del ser humano.
  </p>
  <p>
     Uno de los grandes impulsores actuales son el Big Data, gracias a este campo la IA recopila grandes cantidades de datos para luego ser procesadas y ello determinar patrones y tendencias que serviran para formular predicciones.
  </p>

 <h6><b>¿ cómo funciona la IA ?</b></h6>
  <p>
     Dependiendo el campo de la IA en ocasiones cuenta con diferencias en cuanto a su funcionamiento, sin embargo, las principales técnicas son:
  </p>
  <ul>
     <li><b>Análisis predictivo: </b>
      Establece modelos de puntuación que representa la probabilidad de que algo suceda mediante el análisis de determinadas variables y resultados.
     </li>
     <li><b>Aprendizaje automático:</b>
      Comúnmente es reconocido como machine learning (ML), tiene como objetivo determinar acciones de forma automática sin necesidad de programación explicita.
     </li>
     <li><b>Aprendizaje profundo: </b>
      Se considera como una de las ramas de aprendizaje automático. Estos sistemas de aprendizaje profundo son entrenados a través de ejemplos preexistentes en donde se pueden combinar algoritmos del aprendizaje automático que utilicen redes neuronales complejas, las cuales ayudarán a aprender de los ejemplos o experiencias.
     </li>
     <li><b>Data mining: </b> La mineria de datos permite extraer información oculta de los datos manejados, mediante la exploración y análisis de los datos.</li>
     <li><b>Redes semánticas:</b> Son una representación del conocimiento lingüístico en la que los conceptos e interrelaciones entre ellos son representados por grafos.</li>

  </ul>
